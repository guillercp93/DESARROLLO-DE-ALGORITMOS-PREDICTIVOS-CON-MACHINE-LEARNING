**Hacer limpieza de datos, se debe explicar que metodología se usó para transformar las variables
   y dejar los datos listos. Ademas de resumir el estado original de las variables
   (cantidad de atipicos, cantidad de datos faltantes etc).**

Para las variables categóricas, se hace un reemplazo de los valores faltantes de acuerdo al contexto de la variable.
Es decir, para la columna Alley(Callejón) sería reemplazar el valor N/A por un valor de "No Alley"(Sin callejón).

*Aunque para esas variables con un alto número de variables faltantes, puede considerarse eliminarles del conjunto de datos, pero como no sabemos cuál es la pregunta a resolver, estas columnas pueden tener información que se pueda necesitar a posteriori*

```python
df['Alley'] = df['Alley'].fillna('No Alley')
df['BsmtCond'] = df['BsmtCond'].fillna('No Basement')
df['BsmtExposure'] = df['BsmtExposure'].fillna('No Basement')
df['BsmtFinType1'] = df['BsmtFinType1'].fillna('No Basement')
df['BsmtFinType2'] = df['BsmtFinType2'].fillna('No Basement')
df['BsmtQual'] = df['BsmtQual'].fillna('No Basement')
df['Fence'] = df['Fence'].fillna('No Fence')
df['FireplaceQu'] = df['FireplaceQu'].fillna('No Fireplace')
df['GarageCond'] = df['GarageCond'].fillna('No Garage')
df['GarageFinish'] = df['GarageFinish'].fillna('No Garage')
df['GarageQual'] = df['GarageQual'].fillna('No Garage')
df['GarageType'] = df['GarageType'].fillna('No Garage')
df['MiscFeature'] = df['MiscFeature'].fillna('None')
df['PoolQC'] = df['PoolQC'].fillna('No Pool')
```

Para las variables numéricas con valores faltantes, lo que se hizo fue reemplazarlos con el valor de la media, esto con el fin de evitar un sesgo considerable

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean())
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
```
